```python
import numpy as np
import pandas as pd

def find_financial_indicators(stock_name, year, indicators): # 종목, 년도, 수집할 지표 목록을 받아, 주요 재무지표를 반환
  report = dart.finstate(stock_name, year) # 데이터 가져오기
  if report is None: # 리포트가 없다면 (리포트가 없으면 None 반환)
    # 리포트가 없으면 당기, 전기, 전전기 값 모두 제거
    data = [[stock_name, year] + [np.nan] * len(indicators)]
    data = [[stock_name, year-1] + [np.nan] * len(indicators)]
    data = [[stock_name, year-2] + [np.nan] * len(indicators)]
    return pd.DataFrame(data, columns = ["기업", "연도"] + indicators)

  else:
    report = report[report['account_nm'].isin(indicators)] # 관련 지표로 필터링
    if sum(report['fs_nm'] == "연결재무제표") > 0:
      # 연결재무제표 데이터가 있으면 연결재무제표를 사용
      report = report.loc[report['fs_nm'] == '연결재무제표']
    else:
      # 연결재무제표 데이터가 없으면 일반재무제표를 사용
      report = report.loc[report['fs_nm'] == "재무제표"] 
    
    data = []
    for y, c in zip([year, year-1, year-2], ['thstrm_amount', 'frmtrm_amount', 'bfefrmtrm_amount']):
      record = [stock_name, y]
      for indic in indicators:
        # account_nm이 indic인 행의 c 컬럼 값을 가져옴
        if sum(report['account_nm'] == indic) > 0:
          value = report.loc[report['account_nm'] == indic, c].iloc[0]
        else:
          if(indic == '영업활동 현금흐름'):
            fs_all = dart.finstate_all(stock_name, year) # 제무제표 전체
            fs_all = fs_all.loc[fs_all['sj_nm'] == '현금흐름표'] # 현금흐름표
            value = fs_all.loc[fs_all['account_nm'] == indic, c].iloc[0]
          else:
            value = np.nan

        record.append(value)
      data.append(record)
    return pd.DataFrame(data, columns = ["기업", "연도"] + indicators)
```

```python
# 지표들이 숫자가 아닌 문자로 되어 있기 때문에
# 숫자로 바꿔주는 작업
def str_to_float(value):
  if type(value) == float: # NaN의 경우 문자가 아니라, 이미 float 취급됨
    return value
  elif value == '-': # - 로 되어 있으면 0으로 변환
    return 0
  else:
    return float(value.replace(',',''))

for indc in indicators:
  data[indc] = data[indc].apply(str_to_float)
```

```python
# 부채 비율 계산
# 부채비율 = 자본 총계 / 자산 총계

data['부채비율'] = data['자본총계']/data['자산총계']*100
display(data['부채비율'].head())
```

```python
# 영업이익 증가율 = (당기 영업이익 - 전기영업이익) / 전기 영업이익 * 100%

# 코드로는 기업과 연도를 기준으로 정렬한 뒤, diff 함수를 사용하여 (당기 영업이익 - 전기 영업이익)을 계산
# 단 2013년의 경우 위의 행이 없거나, 있더라도 다른 종목의 2020년도 영업이익이기 때문에 결측으로 바꿔줌

data.sort_values(by = ['기업', '연도'], inplace = True) # 기업과 연도를 기준으로 정렬
data['영업이익증가율'] = data['영업이익'].diff() / data['영업이익']
data.loc[data['연도'] == 2013, '영업이익증가율'] = np.nan
```

```python
# 매출액과 당기 순이익도 같은 방식으로 구현

data.sort_values(by = ['기업', '연도'], inplace = True) # 기업과 연도를 기준으로 정렬
data['매출액증가율'] = data['매출액'].diff() / data['매출액']
data.loc[data['연도'] == 2013, '매출액증가율'] = np.nan

data.sort_values(by = ['기업', '연도'], inplace = True) # 기업과 연도를 기준으로 정렬
data['당기순이익증가율'] = data['당기순이익'].diff() / data['당기순이익']
data.loc[data['연도'] == 2013, '당기순이익증가율'] = np.nan
```

```python
# ROA (Return on assets)
# ROA = (당기순이익/자산총계)*100%

data['ROA'] = data['당기순이익'] / data['자산총계']
```

```python
# ROE (Return on equity)
# ROE = (당기순이익/평균 자기 자본)

average_equity = data['자본총계'].rolling(2).mean() #평균 자기 자본
data['ROE'] = data['당기순이익']/average_equity
```

```python
# 2020년 기준 f_score

def get_f_score(data, company):
  company_data = data.loc[data['기업'] == company]
  factor_1 = int(company_data.loc[data['연도'] == 2020]['ROA']>0) # 현재(2020) ROA가 양수인지
  factor_2 = int(company_data.loc[data['연도'] == 2020]['영업활동 현금흐름']>0) # 현재(2020) CFO가 양수인지
  factor_3 = int(int(company_data.loc[data['연도'] == 2020]['ROA']) > int(company_data.loc[data['연도'] == 2019]['ROA'])) # 현재(2020) ROA가 전년도 ROA보다 큰지
  factor_4 = int(company_data.loc[data['연도'] == 2020]['영업활동 현금흐름']>company_data.loc[data['연도'] == 2020]['ROA']) # 현재 CFO가 ROA보다 큰지
  factor_5 = int(int(company_data.loc[data['연도'] == 2020]['부채비율']) < int(company_data.loc[data['연도'] == 2019]['부채비율'])) # 현재 부채비율이 작년보다 작은지
  # 유동비율 전년대비 증가인지
  factor_6 = int(int(company_data.loc[data['연도'] == 2020]['자산총계']/company_data.loc[data['연도'] == 2020]['부채총계']) > int(company_data.loc[data['연도'] == 2019]['자산총계']/company_data.loc[data['연도'] == 2019]['부채총계']))
  factor_7 = 0 # 전년도 신주 발행 여부
  # 매출총이익률 전년대비 증가인지
  factor_8 = int(int(company_data.loc[data['연도'] == 2020]['당기순이익']/company_data.loc[data['연도'] == 2020]['자산총계']) > int(company_data.loc[data['연도'] == 2019]['당기순이익']/company_data.loc[data['연도'] == 2019]['자산총계']))
  factor_9 = 0 # 회전율 증가
  total_score = factor_1+factor_2+factor_3+factor_4+factor_5+factor_6+factor_7+factor_8+factor_9
  return total_score

get_f_score(data, 'AJ네트웍스')
```

F-Score (Piotroski Score) - \
2000년 스탠퍼드 대학 경영 대학원 회계 교수 조셉 피오트로스키가 2000년에 발표한 가치 투자 연구를 통한 기업 평가 방식 \
\
총 9개의 항목으로 각 1점 점수, 최대 9점 \
\
수익성:
- 순이익(ROA) 흑자
- 영업활동현금흐름(CFO) 흑자
- 현재 ROA > 전년 ROA 
- 영업현금흐름(CFO) > 순이익(ROA)

\
안정성(재무건전성):
- 부채비율 < 전년부채비율 (LEVER 감소)
- 유동비율 > 전년유동비율 (LIQUID 증가)
- 작년 신주발행 이력 없음 (EQ_OFFER) \

\
효율성:
- 매출총이익률 전년대비 증가 (MARGIN 증가)
- 자산회전율 전년대비 증가 (TURN 증가)
